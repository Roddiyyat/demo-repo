### Analysing our dataset.
When i have a data i want to analyse, i tell chat-gpt to come up with a set of questions i can ask my dataset and get answers to. 

Here are a few questions we can solve using our dataset.

1. What is the average salary across the entire dataset?
2. How does salaries varies by job title or position?
3. Are there any notable disparities based on gender or other demographics?
4. How does salaries compare across different geographical location?
5. What are the trends in salary increment over the year?



```python
#What is the average salary across the entire dataset?
my_df['salary'].mean() 
```

```python

# How does salaries varies by job title or position?
my_df.groupby(['job_title', 'experience_level'])['salary'].mean()
my_df.groupby(['job_title', 'experience_level'])['salary'].sum()
```

```python
# How does salaries compare across different geographical location?
my_df.groupby('company_location')['salary'].mean()
my_df.groupby('company_location')['salary'].max()
my_df.groupby('company_location')['salary'].min()
my_df.groupby('company_location')['salary'].median()

```

```python
# What are the trends in salary increment over the year?
# Calculate the average salary for each year
average_salary_per_year = my_df.groupby('work_year')['salary'].mean()

# Calculate the median salary for each year
median_salary_per_year = my_df.groupby('work_year')['salary'].median()

# Calculate the maximum salary for each year
max_salary_per_year = my_df.groupby('work_year')['salary'].max()

# Calculate the minimum salary for each year
min_salary_per_year = my_df.groupby('work_year')['salary'].min()


```
